#### Nettoyage des données:
##### Seul les bâtiments non destinées à l'habitation seront concernées dans ce projet:

```python
df['BuildingType'].unique()
```

```python
df0 = df[~df['BuildingType'].str.contains('Multifamily')]
```

```python
#Vérification:
df0['BuildingType'].unique()
```

```python
df0.shape
```

Le nombre de lignes a été réduit de presque 50%.

##### Vérification des doublons sur l'identifiant OSEBuildingID qui doit être unique

```python
df0.duplicated(subset=['OSEBuildingID']).sum()
```

les identifiants (OSEBuildingID) sont uniques.

On remarque qu'il y a des features qui sont similaires donc on peut conserver les doublons. Par exemple:

SiteEUI(kBtu/sf) & SiteEUIWN(kBtu/sf)               
SourceEUI(kBtu/sf) & SourceEUIWN(kBtu/sf)            
SiteEnergyUse(kBtu) & SiteEnergyUseWN(kBtu)

WN correspond à des données normalisées par rapport aux variations de températures d'une année à l'autre. Donc on peut faire le choix de conserver ces données avec WN. 

```python
# On écarte les features non WN:
df1= df0.drop(['SiteEUI(kBtu/sf)','SourceEUI(kBtu/sf)', 'SiteEnergyUse(kBtu)'],axis=1)
```

```python
#Vérification
df1.columns
```

On remarque également des features identiques avec différentes unités. Par exemple:

Electricity(kWh) & Electricity(kBtu)

NaturalGas(therms) & NaturalGas(kBtu)

(kBtu): the kilo-British thermal unit, is the recommended global unit of measure because it is the one most frequently used in building benchmarks and all ENERGY STAR reporting.

Donc on sélectionne les features avec kBtu comme unité.

```python
# On écarte les features autre que kBtu comme unité :
df2= df1.drop(['Electricity(kWh)','NaturalGas(therms)'],axis=1)
```

```python
# Vérification:
df2.columns
```

Les variables à prédire dans le cadre de notre projet sont la consommation d'énergie du bâtiment (SiteEnergyUseWN(kBtu)) et ses émissions de CO2 (TotalGHGEmissions). 

Ces variables comportent des données manquantes, on va donc supprimer les lignes correspondantes.

```python
df2['SiteEnergyUseWN(kBtu)'].isnull().sum()
```

```python
df2['TotalGHGEmissions'].isnull().sum()
```

```python
# On écarte les lignes avec les données manqauntes:
df3= df2[~((df2['SiteEnergyUseWN(kBtu)'].isnull()) | (df2['TotalGHGEmissions'].isnull()))]
```

```python
#Vérification
df3['SiteEnergyUseWN(kBtu)'].isnull().sum()
```

```python
#Vérification
df3['TotalGHGEmissions'].isnull().sum()
```

On écarte les colonnes qui ont un taux de remplissage très faible (seuil 50%). On a:

ThirdLargestPropertyUseType: 18%

ThirdLargestPropertyUseTypeGFA: 18%

YearsENERGYSTARCertified: 4%

Comments: 0%

Outlier: 0%



```python
df4= df3.drop(['ThirdLargestPropertyUseType','ThirdLargestPropertyUseTypeGFA','YearsENERGYSTARCertified','Comments','Outlier'],axis=1)
```

```python
#Vérfication:
df4.columns
```

#### Feature engineering

##### Analyse des types de bâtiments

```python
df4['BuildingType'].unique()
```

SPS: Seattle Public School

COS: Coefficient d'Occupation du Sol. Le COS est différent d'une zone de la commune à l'autre. 
Un coefficient de 1 signifie que la totalité du terrain peut être utilisée pour la construction. 
Un COS de 0,4 indique que l'on peut utiliser 400 m² de surface de plancher sur un terrain constructible de 1 000 m².

Campus: Vaste terrain construit de bâtiments universitaires et de résidences étudiantes, aux allures de parc, aux États-Unis et au Canada.

```python
# On visualise la répartition des types de bâtiment dans (BuildingType):
building_type = df4.groupby(by='BuildingType')['OSEBuildingID'].nunique()
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(building_type.values, labels=building_type.index,autopct='%1.1f%%')
```

On remarque que les bâtiments de type (NonResidential) représentent la majorité des bâtiments (87.5%).
On va donc s'intéresser plus partuculièrements aux catégories des bâtiments de type (NonResidential).

##### Corrélation linéaire entre les variables:

```python
corr=df4.corr()
```

```python
corr
```

```python
# Heatmap de corrélation linéraire entre les variables:
fig = plt.figure(figsize=(22,8))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr,annot=True,mask=mask,cmap='BrBG',vmin=-1, vmax=1)
```

Rappelons que les variables target de prédictions sont: (TotalGHGEmissions) et (SiteEnergyUseWN(kBtu)).

On identifie les variables par paires qui sont fortement corrélées entre elles. 

On peut définir 0.7 comme seuil de cette forte corrélation.

```python
corr.unstack()
```

```python
corr_paires= corr.unstack().sort_values(ascending=False)
```

```python
corr_paires
```

```python
seuil_corr= 0.7
forte_corr= pd.DataFrame(corr_paires[(abs(corr_paires) > seuil_corr)])
```
