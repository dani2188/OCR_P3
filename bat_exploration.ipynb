A savoir: 

WN: Weather Normalized energy is the energy your property would have used if you had experienced 30-year average temperatures*. The weather in a given year may be much hotter or colder than your building's normal climate; weather normalized energy accounts for this difference

GHGE: greenhouse gas emissions

### Feature engineering et exploration de données

##### Analyse des types de bâtiments

```python
df4['BuildingType'].unique()
```

SPS: Seattle Public School

COS: Coefficient d'Occupation du Sol. Le COS est différent d'une zone de la commune à l'autre. 
Un coefficient de 1 signifie que la totalité du terrain peut être utilisée pour la construction. 
Un COS de 0,4 indique que l'on peut utiliser 400 m² de surface de plancher sur un terrain constructible de 1 000 m².

Campus: Vaste terrain construit de bâtiments universitaires et de résidences étudiantes, aux allures de parc, aux États-Unis et au Canada.

```python
# On visualise la répartition des types de bâtiment dans (BuildingType):
building_type = df4.groupby(by='BuildingType')['OSEBuildingID'].nunique()
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(building_type.values, labels=building_type.index,autopct='%1.1f%%')
```

Filtrage sur les outliers: On va écarter les valeurs oultliers vérifiant:

value > Q3+ 1.5* IQR Or value < Q1 - 1.5* IQR

On écarte les outliers de la variable: (TotalGHGEmissions)

```python
q3_em= df6['TotalGHGEmissions'].quantile(0.75)
q1_em= df6['TotalGHGEmissions'].quantile(0.25)
from scipy.stats import iqr
iqr_em= iqr(df6['TotalGHGEmissions'])
```

```python
df7= df6[~ ((df6['TotalGHGEmissions'] > (q3_em + 1.5*iqr_em)) | (df6['TotalGHGEmissions'] < (q1_em - 1.5*iqr_em)))]
```

```python
#Vérification:
plt.figure(figsize=(5,3))
df7['TotalGHGEmissions'].plot.box(showmeans=True)
```

```python
df7[df7['TotalGHGEmissions']<0]
```

On écarte la ligne qui correspond à la donnée aberrante (TotalGHGEmissions)<0

```python
df8= df7.drop(df7[df7['TotalGHGEmissions']<0].index)
```

```python
# Vérification:
df8[df8['TotalGHGEmissions']<0]
```
